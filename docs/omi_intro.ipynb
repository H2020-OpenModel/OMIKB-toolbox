# Introduction to OMI
- The `Openmodel Open Innovation Infrastructure`
- Provides Users a Coherent Environment to Build, Execute and Curate Workflows
- Central (cloud) and Distributed (in premise, private cloud) installations possible
- Supports OpenModel ExecFlow, OntoFlow and OMIKB
- Additional tools: AiiDa, LAMMPS, and more can be installed. 
- Based on JupyterHUB and uses Keycloak for Authentication 

# The Architecture 
-The back-end provides the interoperability platform, wrappers (to external party data sources, models and simulation software), data storage (triplestore and dataspace) and system monitoring & logging.

- Installing steps
  - edit the .env file for your domain name and passwords etc
  - if using Cloudflare tunnel, a tunnel secret key
  - git clone the repo
  - do `docker compose build` then `docker compose up -d` 
  - configure keycloak and add users]

# Configuration and adding packages
- adding new services can be done in various ways 
    1. adding an end point in omi.py and a containerised service behind omi
    2. adding a containerised service with own authentication against keycloak (and adding it as a client) 
    3. adding it as part of omi (as /ontoflow) 
    4. integrating into the hub as a native app (like execflow and ontoflow, tripper etc), e.g. using pip as a user or globally by creating new docker image for the notebooks
    
    See repo for more information
